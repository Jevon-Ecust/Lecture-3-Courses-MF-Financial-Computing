```python
import numpy as np#numpy
import matplotlib.pyplot as plt#matplotlib
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
#-+256
C,S = np.cos(X), np.sin(X)
#xCS
#xC
plt.plot(X,C)
#xS
plt.plot(X,S)
plt.xticks( [-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
#+-,+-/2 plt.yticks([-1, 0, +1])#+-1,0
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
# 
```
